# Detect patterns in ship behavior based on ship speed and lever position to optimize navigation strategies and improve fuel efficiency.

```python
# lets import the warnings library so that we can avoid warnings
import warnings
warnings.filterwarnings('ignore')

# Lets select the Spending score, and Annual Income Columns from the Data
x_1 = df.loc[:, ['lever_position', 'ship_speed']].values

# let's check the shape of x
print(x_1.shape)
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    km = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    km.fit(x_1)
    wcss.append(km.inertia_)

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method', fontsize = 20)
plt.xlabel('No. of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
plt.style.use('fivethirtyeight')

km = KMeans(n_clusters = 2, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_means = km.fit_predict(x_1)

plt.scatter(x_1[y_means == 0, 0], x_1[y_means == 0, 1], s = 100, c = 'pink', label = 'represents high ship speeds')
plt.scatter(x_1[y_means == 1, 0], x_1[y_means == 1, 1], s = 100, c = 'yellow', label = 'represents low ship speeds')
plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:, 1], s = 50, c = 'blue' , label = 'centeroid')

plt.style.use('fivethirtyeight')
plt.title('K Means Clustering between ship speed and lever position' , fontsize = 20)
plt.xlabel('lever_position')
plt.ylabel('ship_speed')
plt.legend()
plt.grid()
plt.show()

from sklearn.metrics import silhouette_score
# Calculate silhouette score
silhouette_avg = silhouette_score(x_1, y_means)
print("Silhouette Score:", silhouette_avg)
```

# Identify abnormal operating conditions of a ship's gas turbine engine using clustering. This will help with early detection of potential problems and improve preventative maintenance strategies.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler



# Select relevant features for clustering
features = ['gt_shaft_torque', 'gt_rate_of_revolutions', 
            'hp_turbine_exit_temperature', 'gt_compressor_inlet_air_temperature', 
            'gt_compressor_outlet_air_temperature', 'hp_turbine_exit_pressure', 
            'gt_compressor_inlet_air_pressure', 'gt_compressor_outlet_air_pressure', 
            'gt_exhaust_gas_pressure', 'turbine_injection_control', 'fuel_flow']

# Preprocess the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features])

# Perform K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)
df['cluster'] = kmeans.fit_predict(df_scaled)

# Plot the clusters
plt.scatter(df['gt_shaft_torque'], df['gt_rate_of_revolutions'], c=df['cluster'], cmap='viridis')
plt.xlabel('GT Shaft Torque')
plt.ylabel('GT Rate of Revolutions')
plt.title('Clustering of Gas Turbine Engine Operating Conditions')
plt.colorbar(label='Cluster')
plt.show()

# Calculate silhouette score
silhouette_avg = silhouette_score(df_scaled, df['cluster'])

print("Silhouette Score:", silhouette_avg)
```

### In the above cluster, yellow points represent operating conditions where the engine is not heavily loaded. Both torque and revolutions are relatively low.
###  Purple points are scattered at higher values on both axes, these points signify more intense operation. The engine is working harder, with higher torque and faster revolutions.

# To optimize fuel efficiency and ensure safe operation of naval vessels, it's crucial to identify patterns in ship speed and fuel flow data that may indicate abnormal operating conditions. Using clustering, we aim to segment the data into meaningful groups, enabling early detection of anomalies and informing proactive maintenance strategies.


```python
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

# Select relevant features for clustering
features = ['ship_speed', 'fuel_flow']
X = df[features]
```

```python
# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Perform Agglomerative Clustering
agg_cluster = AgglomerativeClustering(n_clusters=2, linkage='ward')
df['cluster'] = agg_cluster.fit_predict(X_scaled)
```

```python
# Visualize Dendrogram
plt.figure(figsize=(10, 6))
dendrogram(linkage(X_scaled, method='ward'))
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
import seaborn as sns
# Visualize Clusters
sns.scatterplot(data=df, x='ship_speed', y='fuel_flow', hue='cluster', palette='viridis')
plt.title('Agglomerative Clustering of Ship Speed vs Fuel Flow')
plt.xlabel('Ship Speed')
plt.ylabel('Fuel Flow')
plt.show()
```

## Cluster 0 (green): Concentrated at the lower end of the ship speed axis, indicating low fuel flow at low speeds.
